```python
from PIL import Image, ImageDraw
from shapely.geometry import Polygon, Point
from shapely.affinity import scale
```

```python
# points of a polygon with 5 edges
points = [(50, 50), (150, 50), (150, 150), (100, 120), (50, 150)]

img = Image.new("RGB", (200, 200), (255, 255, 255))
draw = ImageDraw.Draw(img)
```

```python
# use shapely to create a polygon and buffer it
polygon = Polygon(points)
buffered_polygon = polygon.buffer(10, join_style="mitre")  # buffer by 2 pixels

# get the exterior coordinates of the buffered polygon
buffered_points = list(buffered_polygon.exterior.coords)

# draw the original polygon in red
draw.polygon(points, outline="red", fill=None)
# draw the buffered polygon in blue
draw.polygon(buffered_points, outline="blue", fill=None)

img.show()
```

```python
# ellipse cases

# center
cx, cy = 100, 100

# radius
rx, ry = 50, 30

circle = Point(cx, cy).buffer(1, resolution=64)  # create a unit circle

ellipse = scale(circle, rx, ry)  # scale it to the desired radii

# buffer ecllipse by 10 pixels
buffered_ellipse = ellipse.buffer(10, join_style="round")

img = Image.new("RGB", (200, 200), (255, 255, 255))
draw = ImageDraw.Draw(img)

# draw the original ellipse in red
draw.ellipse((cx - rx, cy - ry, cx + rx, cy + ry), outline="red", fill=None)
# draw the buffered ellipse in blue
buffered_points = list(buffered_ellipse.exterior.coords)
draw.polygon(buffered_points, outline="blue", fill=None)

# try to buffer the ellipse by adding 10 directly to the radii
rx_buffered = rx + 5
ry_buffered = ry + 5
draw.ellipse((cx - rx_buffered, cy - ry_buffered, cx + rx_buffered, cy + ry_buffered), outline="green", fill=None)
img.show()
```

```python
# reactangle cases
# top-left corner
x, y = 50, 50
width, height = 100, 60

img = Image.new("RGB", (200, 200), (255, 255, 255))
draw = ImageDraw.Draw(img)

# draw the original rectangle in red
draw.rectangle((x, y, x + width, y + height), outline="red", fill=None)

# buffer the rectangle by 10 pixels
buffer_size = 10
buffered_x = x - buffer_size
buffered_y = y - buffer_size
buffered_width = width + 2 * buffer_size
buffered_height = height + 2 * buffer_size
draw.rectangle((buffered_x, buffered_y, buffered_x + buffered_width, buffered_y + buffered_height), outline="blue", fill=None)

img.show()
```
